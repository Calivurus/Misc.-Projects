```python
import random

class Coin:
    def __init__(self, rare = False, clean = True, heads = True, **kwargs):
        
        for key,value in kwargs.items():
            setattr(self,key,value)
        
        self.is_rare = rare
        self.is_clean = clean
        self.heads = heads
        
        
        if self.is_rare:
            self.value = self.original_value *1.25
        else:
            self.value = self.original_value

        if self.is_clean:
            self.color = self.clean_color
        else:
            self.color = self.rusty_color 
            
    def rust(self):
        self.color = self.rusty_color  
        
    def clean(self):
        self.color = self.clean_color 
        
    def __del__(self):
        print("Coin spent!")
        
    def flip(self):
        head_options = [True, False]
        choice = random.choice(head_options)
        self.heads = choice
        
    def __str__(self):
        if self.original_value > 1.00:
            return "${} Coin".format(int(self.original_value))
        else:
            return "{}cent Coin".format(int(self.original_value*100))
    
class Sacagawea_Dollar(Coin):
    def __init__(self):
        data = {  
            "original_value": 1.00,
            "clean_color": "gold",
            "rusty_color": "greenish",
            "num_edges": 1,
            "diameter": 26.5, #mm
            "thickness": 2, #mm
            "mass": 8.1 #grams
            }
        super().__init__(**data)
        
class Half_dollar(Coin):
    def __init__(self):
        data = {  
            "original_value": 0.50,
            "clean_color": "Silver",
            "rusty_color": None,
            "num_edges": 1,
            "diameter": 30.61, #mm
            "thickness": 2.15, #mm
            "mass": 11.34 #grams
            }
        super().__init__(**data)
        def rust(self):
            self.color=self.clean_color
        
class Quarter(Coin):
    def __init__(self):
        data = {  
            "original_value": 0.25,
            "clean_color": "Silver",
            "rusty_color": None,
            "num_edges": 1,
            "diameter": 24.26, #mm
            "thickness": 1.75, #mm
            "mass": 5.64 #grams
            }
        super().__init__(**data)
        def rust(self):
            self.color=self.clean_color
        
class Dime(Coin):
    def __init__(self):
        data = {  
            "original_value": 0.10,
            "clean_color": "Silver",
            "rusty_color": None,
            "num_edges": 1,
            "diameter": 17.91, #mm
            "thickness": 1.35, #mm
            "mass": 2.29 #grams
            }
        super().__init__(**data)
        def rust(self):
            self.color=self.clean_color
        
class Nickel(Coin):
    def __init__(self):
        data = {  
            "original_value": 0.05,
            "clean_color": "Silver",
            "rusty_color": None,
            "num_edges": 1,
            "diameter": 21.21, #mm
            "thickness": 1.95, #mm
            "mass": 5 #grams
            }
        super().__init__(**data)
        def rust(self):
            self.color=self.clean_color
        
class Penny(Coin):
    def __init__(self):
        data = {  
            "original_value": 0.01,
            "clean_color": "copper",
            "rusty_color": "greenish",
            "num_edges": 1,
            "diameter": 19.05, #mm
            "thickness": 1.52, #mm
            "mass": 2.5 #grams
            }
        super().__init__(**data)
        
```

```python
coins =[Sacagawea_dollar(),
        Quarter(),
        Dime(),
        Nickel(),
        Penny()
       ]
for coin in coins:
    arguments =[
        coin, coin.color, coin.value,
        coin.diameter, coin.thickness,
        coin.num_edges, coin.mass
        ]
    string = "{} - Color:{}, value:{}, diameter(mm):{}, thickness(mm):{}, number of edges:{}, mass(g):{}".format(*arguments)
    print(string)
```
